```python
import os
import json
#os.chdir('D:/Users/feelus/Miniconda3/pkgs')
os.chdir('C:/ProgramData/Miniconda3/pkgs')
```

```python
l={}
with os.scandir('.') as it:
    for entry in it:
        if not entry.name.startswith('.') and entry.is_dir()\
           and  os.path.isfile(entry.name+'/info/index.json'):
            tmp=json.load(open(entry.name+'/info/index.json'))
            l[tmp['name']]=[y for y in map(lambda s: s.split(' ')[0],tmp['depends'])]
l
```

```python
ln={}
n=1
for key in l:
    ln[key]=n
    n+=1
ln
```

```python
nn=[]
for k,v in l.items():
    for j in v:
        nn.append((ln[k],ln[j]))
nn
```

```python
os.chdir('C:/Users/feelus/Desktop')
fp=open('deps.tgf','w')
for name,num in ln.items():
    print(num,' ',name,file=fp)
print('#',file=fp)
for a,b in nn:
    print(a,' ',b,file=fp)
fp.close()
```
